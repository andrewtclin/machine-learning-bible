## Polynomial Linear Regression

We can view it as the **special case of multiple linear regression**.

<img src='../../resources/polynomial regression/equation.png' />

Similar to linear regression, but:
1. Instead of using different variables of x, we use the same x variable but varies the **power of x with n.**

In the following graph, adding b2x1^2 gives the curve to best fit the dataset.

<img src='../../resources/polynomial regression/graph.png' />

Despite of the graph looks non-linear (x & y non-linear relationship), but we refers `linear` not for the x variable, but for the **coefficient of the slope**.

<hr />

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing Data

```python
df = pd.read_csv('data/position_salaries.csv')
df
```

```python
X = df.drop(['Salary', 'Position'], axis=1)
y = df['Salary']
```

### Training Linear Regression model on the entire data set

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(X, y)
```

### Training Polynomial Regression model on the entire dataset

<img src='../../resources/polynomial regression/equation2.png' />

To build this, we will first create a **matrix for the power**.

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 2) # The power of x
X_poly = poly_reg.fit_transform(X) # Use only x1 - 'Level Column'

poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_poly, y)
```

### Linear Regression Result

```python
plt.scatter(X, y, color='salmon')
plt.plot(X, lin_reg.predict(X), color='lightblue')
plt.title('Position vs Salary (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary');
```

### Polynomial Regression Result

```python
plt.scatter(X, y, color='salmon')
plt.plot(X, poly_lin_reg.predict(X_poly), color='lightblue')
plt.title('Position vs Salary (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary');
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 4) # The power of x, including power of 1, 2, 3, 4
X_poly = poly_reg.fit_transform(X) # Use only x1 - 'Level Column'

poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_poly, y)
```

```python
# Just demonstration, Overfitting

plt.scatter(X, y, color='salmon')
plt.plot(X, poly_lin_reg.predict(X_poly), color='lightblue')
plt.title('Position vs Salary (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary');
```

### Predicting a result with Linear Regression

```python
lin_reg.predict([[6.5]]) # Input level
# If we input [6.5], it is a 1D vector
# [[6.5]] is 2D, 1 row, 1 column
```

### Predicting a result with Polynomial Regression

```python
poly_lin_reg.predict(poly_reg.fit_transform([[6.5]])) # We need to transform feature to fit format of polynomial
```
